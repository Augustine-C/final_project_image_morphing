```python
import cv2
import glob
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.tri as tri
from skimage import draw
import os
```

```python
def cross_dissolve(img_1, img_2, duration_in_second, fps=30):
    num_frames = duration_in_second * fps
    output_subdir = 'cross_dissolve'
    os.makedirs('./images/outputs/{}/'.format(output_subdir), exist_ok = True)

    # generate the frames
    for frame in range(num_frames + 1):
        t = frame / num_frames
        img_halfway = (1-t) * img_1 + t*img_2
        cv2.imwrite('./images/outputs/{}/frame_{}.jpg'.format(output_subdir, str(frame)), img_halfway)
```

```python
img1 = cv2.imread('./images/inputs/cat.jpg')
img2 = cv2.imread('./images/inputs/tiger.jpg')

%matplotlib inline
fig = plt.figure()
plt.imshow(img1[:,:,[2,1,0]])

fig = plt.figure()
plt.imshow(img2[:,:,[2,1,0]])

duration = 5 # in seconds
fps = 30
num_frames = duration * fps
results = []
cross_dissolve(img1, img2, duration, fps)
```
